 1.What is hypothesis testing in statistics?
 --
 Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population based on sample data. It involves testing a claim or hypothesis about a population parameter to determine whether there is enough evidence to support or reject that claim.

 2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
 --
 The null hypothesis (H₀) and the alternative hypothesis (H₁ or Ha) are two competing statements used in hypothesis testing to make inferences about a population based on sample data.

 The null hypothesis is a statement of no effect, no difference, or no relationship. It represents the status quo or a default position.
In most cases, the null hypothesis suggests that any observed effect in the data is due to random chance rather than a real underlying effect. The goal of hypothesis testing is often to gather evidence to reject the null hypothesis.

The alternative hypothesis is a statement that represents what you want to prove or explore. It suggests that there is an effect, there is a difference, or there is a relationship in the population. The alternative hypothesis is the complement of the null hypothesis; if the null hypothesis is rejected, the alternative hypothesis is considered to be supported.

3.What is the significance level in hypothesis testing, and why is it important?
--
The significance level (denoted as α) in hypothesis testing is a threshold that determines how much evidence you need in order to reject the null hypothesis (H₀). It is the probability of making a Type I error, which is the mistake of rejecting the null hypothesis when it is actually true.
The significance level is a pre-determined value that you set before conducting the test. It represents the risk you’re willing to take in making a Type I error (rejecting a true null hypothesis). Common values for α are 0.05 (5%), 0.01 (1%), and 0.10 (10%). The most commonly used α is 0.05, which means there’s a 5% chance of incorrectly rejecting the null hypothesis.

The significance level serves several critical roles in hypothesis testing:

Determines the Threshold for Rejection

Balances the Risk of Errors

Helps to Make Objective Decisions

Reflects the Strength of Evidence

4 What does a P-value represent in hypothesis testing?
--
The p-value in hypothesis testing is a measure that helps you determine the strength of the evidence against the null hypothesis (H₀). It represents the probability of obtaining test results at least as extreme as the results actually observed, under the assumption that the null hypothesis is true.

The p-value is the probability that you would observe your sample data, or something more extreme, if the null hypothesis were true.
A small p-value indicates that the observed data is unlikely under the null hypothesis, suggesting evidence in favor of the alternative hypothesis (H₁).
A large p-value suggests that the observed data is consistent with the null hypothesis, indicating insufficient evidence to support the alternative hypothesis.

5.How do you interpret the P-value in hypothesis testing?
--
Interpreting the p-value in hypothesis testing is crucial for making decisions about whether to reject or fail to reject the null hypothesis (H₀).

Set a Significance Level (α):
Before performing hypothesis testing, you select a significance level, typically denoted as α. Common values for α are:

0.05 (5%)
0.01 (1%)
0.10 (10%)
This α value is the threshold for deciding whether the evidence is strong enough to reject the null hypothesis.

Compare the p-value to the significance level (α):

If the p-value ≤ α:
You reject the null hypothesis (H₀). The result is considered statistically significant, meaning there is enough evidence to support the alternative hypothesis (H₁).

If the p-value > α:
You fail to reject the null hypothesis (H₀). The result is not statistically significant, meaning there is insufficient evidence to support the alternative hypothesis.

6.What are Type 1 and Type 2 errors in hypothesis testing?
--
In hypothesis testing, there are two types of errors you can make when drawing conclusions from your statistical test:

Type I Error (False Positive):

Definition: A Type I error occurs when you reject the null hypothesis (H₀) when it is actually true.
In other words, you incorrectly conclude that there is an effect or difference when, in reality, there isn't one.
The probability of making a Type I error is denoted by α (the significance level). For example, if α = 0.05, there is a 5% chance of making a Type I error.

Example of a Type I error:
Suppose you are testing whether a new drug lowers blood pressure (null hypothesis: no effect, alternative hypothesis: the drug has an effect).
A Type I error would occur if, based on your sample data, you reject the null hypothesis and conclude that the drug does lower blood pressure, even though it actually does not.

Type II Error (False Negative):
Definition: A Type II error occurs when you fail to reject the null hypothesis (H₀) when it is actually false.In other words, you incorrectly conclude that there is no effect or difference when, in reality, there is one.The probability of making a Type II error is denoted by β. The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.

Example of a Type II error:

Suppose you're testing the same new drug for blood pressure reduction.
A Type II error would occur if, based on your sample data, you fail to reject the null hypothesis (you conclude that the drug has no effect) even though the drug actually does lower blood pressure.

7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
--
In hypothesis testing, a one-tailed test and a two-tailed test refer to the direction of the test and the way you set up the alternative hypothesis (H₁). The main difference between them is how you define the region in which you would reject the null hypothesis based on the data.

One-Tailed Test:

Use when you are only interested in one direction of the effect. For instance, you might use a one-tailed test if you believe a new drug will only increase a variable (e.g., blood pressure), not decrease it.
You should have a strong theoretical or prior reasoning to justify that the effect can only occur in one direction.

Two-Tailed Test:

Use when you are testing for any difference from the null hypothesis, without assuming the direction of the effect.
This is more common, as most hypothesis testing is concerned with whether there is any difference, not just in one direction.

8.What is the Z-test, and when is it used in hypothesis testing?
--
A Z-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between the sample mean and the population mean (or between two sample means) when the population standard deviation is known, and the sample size is large enough. The Z-test uses the Z-distribution (also known as the standard normal distribution), which has a mean of 0 and a standard deviation of 1.

A Z-test is used in the following situations:

When the population standard deviation is known:
The Z-test requires knowledge of the population standard deviation (σ). If you do not know the population standard deviation, you would typically use a t-test instead, especially for small sample sizes.

When the sample size is large (typically n > 30):
The Z-test assumes that the sampling distribution of the sample mean is approximately normal. According to the Central Limit Theorem, if the sample size is large enough (usually n > 30), the sampling distribution will be approximately normal regardless of the shape of the population distribution.

For comparing a sample mean to a population mean (one-sample Z-test):
The one-sample Z-test is used to compare the mean of a sample to a known population mean (or hypothesized population mean).

For comparing two sample means (two-sample Z-test):
A two-sample Z-test is used when comparing the means of two independent samples. This is applicable when the standard deviations of both populations are known.

For proportions (Z-test for proportions):
A Z-test can also be used to test hypotheses about population proportions, particularly in situations like determining whether the proportion of success in a sample is significantly different from a known proportion in the population.

14.What is a confidence interval, and how is it used to interpret statistical results?
--
A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate an unknown population parameter. It gives a range within which we are fairly certain the true population parameter lies, based on the sample data.
`For example, if we are estimating the mean of a population, the confidence interval provides a range of values within which the true population mean is likely to fall, given the data and a specified confidence level.

Estimating Population Parameters: Confidence intervals are often used to estimate unknown population parameters, such as the population mean, population proportion, or population difference between two means.

Hypothesis Testing: When testing a hypothesis about a population parameter, the confidence interval can be used to assess the plausibility of the null hypothesis. For example:

If a hypothesized value (like a population mean) lies outside the confidence interval, you may reject the null hypothesis.
If the hypothesized value lies inside the interval, you may fail to reject the null hypothesis.
Comparing Two Groups: When comparing two groups (e.g., in an A/B test), you can compute confidence intervals for the differences between the means of the two groups. If the confidence interval for the difference contains 0, you cannot reject the null hypothesis that the two groups have the same mean.

Decision-Making: Confidence intervals give decision-makers a range of plausible values for an unknown parameter, which helps inform decisions, especially when uncertainty is involved.

20.What is an ANOVA test, and what are its assumptions?
--
The ANOVA (Analysis of Variance) test is a statistical method used to compare means across three or more groups to determine if there are any statistically significant differences between them. It is a powerful tool that helps assess whether variations within each group are smaller than the variations between the groups. Essentially, it tests whether the group means are equal or if at least one group mean differs significantly from the others.

Assumptions of ANOVA:

ANOVA, like many statistical tests, has specific assumptions that must be met for the results to be valid. Violating these assumptions can lead to incorrect conclusions.

1. Independence of Observations
Assumption: The data points in each group must be independent of each other. This means that the observations within each group should not influence one another.
Why it matters: If data points are not independent, the results of ANOVA can be biased and misleading.
2. Normality
Assumption: The data in each group should follow a normal distribution (or at least be approximately normal).
Why it matters: ANOVA relies on the assumption that the data within each group is normally distributed. This is particularly important when sample sizes are small, as the Central Limit Theorem (CLT) assures normality for larger sample sizes.
Testing normality: You can use tests like the Shapiro-Wilk test or Q-Q plots to assess normality.
3. Homogeneity of Variances (Homoscedasticity)
Assumption: The variances within each group should be approximately equal.
Why it matters: If the variances are not equal across groups, the F-statistic may not follow the F-distribution, and the test could be invalid. Unequal variances can lead to inflated Type I error rates or reduced power.
Testing homogeneity of variances: You can use tests like Levene’s test or Bartlett’s test to assess this assumption.
4. Random Sampling
Assumption: The samples must be randomly selected from the population to avoid bias.
Why it matters: Random sampling ensures that the groups represent the population well and that results are generalizable.


21.What are the different types of ANOVA tests?
--
There are several types of ANOVA (Analysis of Variance) tests used in statistics, each designed to handle different experimental designs or research questions. These tests primarily differ based on the number of factors (independent variables) involved, whether the factors are repeated or not, and the structure of the data. Below are the main types of ANOVA tests:

1. One-Way ANOVA

Purpose:

One-Way ANOVA is used when you have one independent variable (factor) with three or more levels (groups) and want to compare the means of these groups to see if at least one of them is different.

Example:

Testing the effect of different types of diets (low-carb, low-fat, and balanced) on weight loss.

2. Two-Way ANOVA

Purpose:

Two-Way ANOVA is used when you have two independent variables (factors) and want to examine their individual and interaction effects on a dependent variable.

Example:

Investigating how diet type (low-carb, low-fat, balanced) and exercise routine (cardio, weightlifting, none) affect weight loss. Here, diet type and exercise routine are the two independent variables, and weight loss is the dependent variable


3. Repeated Measures ANOVA

Purpose:

Repeated Measures ANOVA is used when the same subjects are measured multiple times under different conditions (i.e., repeated measurements) or over time.
It is useful when the data involve multiple measurements taken on the same subject or unit.

Example:

Testing the effect of a drug on blood pressure at multiple time points (before treatment, 1 week, 2 weeks, 4 weeks, etc.) in the same group of patients.

4. Multivariate Analysis of Variance (MANOVA)

Purpose:

MANOVA is an extension of ANOVA that allows for the analysis of multiple dependent variables simultaneously. This is useful when you have more than one dependent variable and want to test whether the mean vectors for the groups are equal.

Example:

Investigating how different teaching methods affect both student performance (exam scores) and student satisfaction (survey scores).

5. Analysis of Covariance (ANCOVA)

Purpose:

ANCOVA is a combination of ANOVA and regression analysis. It is used when you want to compare one or more group means while controlling for the effect of one or more continuous variables (called covariates).

Example:

Investigating the effect of different study methods on exam scores, while controlling for the number of hours spent studying.

6. Mixed-Design ANOVA (Split-Plot ANOVA)

Purpose:

Mixed-Design ANOVA is used when you have both repeated measures and independent groups. It combines aspects of a repeated measures ANOVA and a two-way ANOVA.

Example:

A study where subjects are assigned to one of three teaching methods (between-subjects factor), and each subject is tested at multiple time points (within-subjects factor) to measure performance.

22.What is the F-test, and how does it relate to hypothesis testing?
--
F-Test and Its Role in Hypothesis Testing

The F-test is a statistical test that compares the variances of two or more groups to determine if they are significantly different from each other. The F-test is commonly used in the context of ANOVA (Analysis of Variance), regression analysis, and testing the equality of variances between two or more groups. It is a way to test hypotheses about the relative sizes of variances or the overall significance of models.

Relation Between F-Test and Hypothesis Testing

In hypothesis testing, the F-test is used to test whether there are significant differences between group variances (in the case of ANOVA) or to assess the overall significance of a model (in the case of regression).

Here is the basic relationship:

Hypothesis Formation: In hypothesis testing, we start by defining a null hypothesis that usually states that there is no effect or no difference (e.g., no difference in means or no significant relationship).

Test Statistic Calculation: The F-test calculates an F-statistic that compares variances (either between groups in ANOVA or explained vs. unexplained variance in regression).

Decision Making: The calculated F-statistic is compared to the critical value from the F-distribution table based on the degrees of freedom (df) and the chosen significance level (typically 0.05). If the F-statistic exceeds the critical value, you reject the null hypothesis.

Conclusion: Based on the comparison between the calculated F-statistic and the critical F-value, you make a decision on whether to reject or fail to reject the null hypothesis.




```python
# 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

# Function to perform Z-test
def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed test

    # Interpret the result
    if p_value < alpha:
        result = f"Reject the null hypothesis. p-value = {p_value:.4f}"
    else:
        result = f"Fail to reject the null hypothesis. p-value = {p_value:.4f}"

    return z_score, p_value, result

# Example parameters
sample_mean = 105  # Sample mean
population_mean = 100  # Known population mean
population_std = 15  # Population standard deviation
sample_size = 30  # Sample size
alpha = 0.05  # Significance level (for 95% confidence)

# Perform the Z-test
z_score, p_value, result = z_test(sample_mean, population_mean, population_std, sample_size, alpha)

# Print the results
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

```python
# 9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

# Function to calculate confidence interval for a sample mean
def calculate_confidence_interval(sample, confidence_level=0.95):
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (using ddof=1 for unbiased estimate)

    # Determine the critical value (z or t) for the desired confidence level
    alpha = 1 - confidence_level
    degrees_of_freedom = sample_size - 1

    # Use T-distribution for smaller sample sizes (n < 30) or if population std dev is unknown
    t_critical = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Confidence interval calculation
    margin_of_error = t_critical * standard_error
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return sample_mean, ci_lower, ci_upper, margin_of_error

# Example usage
np.random.seed(42)  # For reproducibility

# Simulate a sample data (e.g., 30 observations from a normal distribution)
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Calculate confidence interval
sample_mean, ci_lower, ci_upper, margin_of_error = calculate_confidence_interval(sample_data, confidence_level=0.95)

# Output the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Margin of Error: {margin_of_error:.2f}")
```

```python
# 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Bayes' Theorem Implementation in Python

def bayesian_inference(prior_H, likelihood_E_given_H, false_positive_rate, prior_neg_H):

    # P(E) = P(E|H) * P(H) + P(E|¬H) * P(¬H) (total probability of observing the evidence E)
    marginal_likelihood_E = (likelihood_E_given_H * prior_H) + (false_positive_rate * prior_neg_H)

    # Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior_H_given_E = (likelihood_E_given_H * prior_H) / marginal_likelihood_E

    return posterior_H_given_E

# Given probabilities
prior_H = 0.01  # P(H) - Prior probability of having the disease
likelihood_E_given_H = 0.95  # P(E|H) - Probability of testing positive given having the disease
false_positive_rate = 0.05  # P(E|¬H) - Probability of testing positive without the disease (False positive rate)
prior_neg_H = 1 - prior_H  # P(¬H) - Prior probability of not having the disease

# Calculate the posterior probability using Bayes' Theorem
posterior_H_given_E = bayesian_inference(prior_H, likelihood_E_given_H, false_positive_rate, prior_neg_H)

# Output the result
print(f"The posterior probability of having the disease given a positive test result is: {posterior_H_given_E:.4f}")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

# Function to calculate expected frequencies for a Chi-square test
def calculate_expected_frequencies(observed_data):
    # Calculate row totals, column totals, and grand total
    row_totals = np.sum(observed_data, axis=1)  # Sum along rows
    column_totals = np.sum(observed_data, axis=0)  # Sum along columns
    grand_total = np.sum(observed_data)  # Total sum of all cells

    # Calculate expected frequencies using the formula
    expected_data = np.outer(row_totals, column_totals) / grand_total

    return expected_data

# Example observed data (Contingency Table)
# Rows: Categories (e.g., Gender: Male, Female)
# Columns: Categories (e.g., Drink Preference: Tea, Coffee)
observed_data = np.array([[40, 60],  # Male: 40 Tea, 60 Coffee
                          [50, 50]])  # Female: 50 Tea, 50 Coffee

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the observed and expected frequencies
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
# 16.  Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats

# Parameters for the random samples
np.random.seed(42)  # For reproducibility
sample_size = 100  # Sample size for both groups
variance_1 = 4  # Variance of first sample
variance_2 = 9  # Variance of second sample

# Generate random samples
sample_1 = np.random.normal(0, np.sqrt(variance_1), sample_size)
sample_2 = np.random.normal(0, np.sqrt(variance_2), sample_size)

# Calculate sample variances
s1_squared = np.var(sample_1, ddof=1)  # Sample variance for sample 1
s2_squared = np.var(sample_2, ddof=1)  # Sample variance for sample 2

# Calculate F-statistic
F_statistic = s1_squared / s2_squared if s1_squared >= s2_squared else s2_squared / s1_squared

# Degrees of freedom
df1 = sample_size - 1  # Degrees of freedom for sample 1
df2 = sample_size - 1  # Degrees of freedom for sample 2

# Calculate p-value from the F-distribution
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

# Significance level (alpha)
alpha = 0.05

# Output results
print(f"Sample 1 Variance: {s1_squared}")
print(f"Sample 2 Variance: {s2_squared}")
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")

# Hypothesis testing
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

```

```python
# 19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(groups):
    """
    Function to check the assumptions for ANOVA: normality, independence, and equal variance.

    Parameters:
    - groups: List of arrays (each array is a group of data).

    Returns:
    - None. Prints out results for each assumption test.
    """

    # 1. Check Normality Assumption using the Shapiro-Wilk test for each group
    print("Checking normality assumption:")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i} - Shapiro-Wilk Test p-value: {p_value}")
        if p_value < 0.05:
            print(f"Group {i} does NOT follow a normal distribution (p < 0.05).")
        else:
            print(f"Group {i} follows a normal distribution (p >= 0.05).")
    print("\n")

    # 2. Check Independence Assumption: This can't be tested directly through code, assuming data is independent
    print("Checking independence assumption: Assuming the data is independent based on study design.\n")

    # 3. Check Homogeneity of Variance (Equal Variance) using Levene's Test
    print("Checking homogeneity of variance assumption:")
    stat, p_value = stats.levene(*groups)  # Levene's test for equal variances
    print(f"Levene's Test p-value: {p_value}")
    if p_value < 0.05:
        print("The variances are significantly different (p < 0.05). The homogeneity of variance assumption is violated.")
    else:
        print("The variances are not significantly different (p >= 0.05). The homogeneity of variance assumption is not violated.")
    print("\n")

    # 4. Visualize Normality with histograms and Q-Q plots
    print("Visualizing normality assumption:")
    for i, group in enumerate(groups, start=1):
        # Plot histogram
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        plt.hist(group, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
        plt.title(f"Group {i} - Histogram")

        # Q-Q plot
        plt.subplot(1, 2, 2)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Group {i} - Q-Q Plot")
        plt.tight_layout()
        plt.show()


# Example usage
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

groups = [group1, group2, group3]

# Check assumptions
check_anova_assumptions(groups)
```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.anova import anova_lm

# Step 1: Generate data for two-way ANOVA
np.random.seed(42)

# Factor A: 3 levels (A1, A2, A3)
# Factor B: 2 levels (B1, B2)
# Dependent variable: 'value'

factor_A = np.repeat(['A1', 'A2', 'A3'], 20)
factor_B = np.tile(np.repeat(['B1', 'B2'], 10), 3)
values = np.random.normal(loc=50, scale=10, size=60)

# Adding some effects (for illustration)
values[::3] += 5  # A1 has higher values
values[1::3] -= 5  # A2 has lower values

# Create a DataFrame
data = pd.DataFrame({
    'Factor_A': factor_A,
    'Factor_B': factor_B,
    'Value': values
})

# Step 2: Perform two-way ANOVA using statsmodels
model = smf.ols('Value ~ Factor_A * Factor_B', data=data).fit()  # Factor_A * Factor_B tests interaction
anova_table = anova_lm(model)
print(anova_table)

# Step 3: Visualization of the interaction
sns.set(style="whitegrid")

# Interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(x="Factor_A", y="Value", hue="Factor_B", data=data, dodge=True, markers=["o", "s"], linestyles=["-", "--"])
plt.title('Two-Way ANOVA Interaction Plot')
plt.show()
```

```python
# 24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility

# Sample data: Let's say we have 30 observations with a true population variance of 100
sample_data = np.random.normal(loc=50, scale=10, size=30)

# Step 2: Calculate sample variance
sample_variance = np.var(sample_data, ddof=1)  # ddof=1 for sample variance

# Step 3: Set hypothesized population variance (σ₀²)
hypothesized_variance = 100

# Step 4: Calculate the Chi-square test statistic
n = len(sample_data)
chi_square_stat = (n - 1) * sample_variance / hypothesized_variance

# Step 5: Determine the degrees of freedom
df = n - 1

# Step 6: Compute the p-value (two-tailed test)
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Step 7: Output the test statistic and p-value
print(f"Chi-square Test Statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Step 8: Interpretation of the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The sample variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the hypothesized variance.")
```

```python
# 25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import scipy.stats as stats

# Step 1: Sample data (successes and sample sizes for both groups)
x1 = 120  # Number of successes in group 1
n1 = 200  # Sample size for group 1

x2 = 130  # Number of successes in group 2
n2 = 220  # Sample size for group 2

# Step 2: Calculate the sample proportions
p1 = x1 / n1  # Proportion of successes in group 1
p2 = x2 / n2  # Proportion of successes in group 2

# Step 3: Calculate the pooled proportion (P)
P = (x1 + x2) / (n1 + n2)

# Step 4: Calculate the Z-test statistic
z_statistic = (p1 - p2) / np.sqrt(P * (1 - P) * (1 / n1 + 1 / n2))

# Step 5: Compute the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

# Step 6: Output the test statistic and p-value
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Step 7: Interpretation of the results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")

```
